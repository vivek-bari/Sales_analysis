```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Amazon Sale Report.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['New','PendingS'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
# to save the changes write inplace='True'
df['ship-postal-code']=df['ship-postal-code'].astype('int')
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.rename(columns={'Qty':'Quantity'})
```

```python
df.describe()
```

```python
ax=sns.countplot(x='Size',data=df)
```

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
s_Qty = df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
sns.barplot(x='Size', y='Quantity', data=s_Qty)
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x='Courier Status',hue='Status')
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

```python
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title("Distribution of Satte")
plt.xticks(rotation=90)
plt.show()
```

```python
top_states = df['ship-state'].value_counts().nlargest(10).reset_index()
top_states.columns = ['ship-state', 'count']
plt.figure(figsize=(12,6))
sns.barplot(data=top_states, x='ship-state', y='count', palette='viridis')
plt.xlabel('Ship State')
plt.ylabel('Order Count')
plt.title('Top 10 Shipping States by Order Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_states = df['ship-state'].value_counts().nlargest(10).reset_index()
top_states.columns = ['ship-state', 'count']

plt.figure(figsize=(12,6))
ax = sns.barplot(data=top_states, x='ship-state', y='count', palette='viridis')
plt.xlabel('Ship State')
plt.ylabel('Order Count')
plt.title('Top 10 Shipping States by Order Count')
plt.xticks(rotation=45)

for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```
